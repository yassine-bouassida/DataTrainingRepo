```python
import re
#https://docs.python.org/3/library/re.html
states = [' Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda','south carolina##', 'West virginia?']
def clean_strings(strings):
    result = []
    for value in strings:
        #removes white space
        value=value.strip()
        value=re.sub('[!#?]','',value)
        value=value.title()
        result.append(value)
    return result

clean_strings(states)
```

```python
def remove_punctuation(value):
    #return re.sub('or|[ab]','',value,flags=re.I)
    return re.sub('[?!#]','',value,flags=re.I)

clean_ops=[str.strip,remove_punctuation,str.title]
def clean_strings(strings, ops):
    result=[]
    #value below is a string and strings is the list of strings
    for value in strings:
        for function in ops:
            value=function(value)
        result.append(value)
    return result

clean_strings(states,clean_ops)
```

```python
y=[x for x in map(remove_punctuation,states)]
[z for z in map(str.title,y)]
```

```python
#generators
def squares(n=10):
    print('Generating squares from 1 to {0}'.format(n**2))
    for i in range(1,n+1):
        print(i)
        yield i**2
```

```python
gen=squares(10000)
```

```python
for x in gen:
    print(x,end=' ')
```

```python
#another even more concise way to use generators is generator expressions
gen = (x ** 2 for x in range(10000000))
#next(gen)
#gen.__next__()
# next(gen)
# next(gen)
```

```python
for value in gen:
    if value>100000:
        break
    elif value>1000:
        print(value)
    else:
        pass
```

```python
import itertools
L=[("a",1),("a",2),("b",3),("b",4)]
key_func=lambda x: x[0]

some_dict={}
for key, group in itertools.groupby(L,key_func):
    #print(key + ":",list(group))
    some_dict[key]=list(group)

some_dict
```

```python
# example of iterators
type(some_dict)
```

```python
dict_iterator=iter(some_dict)
type(dict_iterator)
next(dict_iterator)
```
