```python
# Delete rows with empty cells.
df = pd.read_csv("dirtydata.csv")
print(df.to_string())

print("After dropping...")
new_df = df.dropna()
print(new_df.to_string())
```

```python
df.dropna(inplace = True)
print(df.to_string())
```

```python
# Replace empty values.
df = pd.read_csv("dirtydata.csv")
print(df.to_string())

print("After replacing...")
df.fillna(130, inplace = True)
print(df.to_string())
```

```python
# Replace empty values in specific Series (column).
df = pd.read_csv("dirtydata.csv")
print(df.to_string())

print("After replacing...")
df["Calories"]=df["Calories"].fillna(130)
print(df.to_string())
```

```python
# Replace empty values with mean() in specific Series (column).
df = pd.read_csv("dirtydata.csv")
print(df.to_string())

print("After replacing...")
calories_mean = df["Calories"].mean()
df["Calories"]=df["Calories"].fillna(calories_mean)

print(df.to_string())
```

```python
# Replace empty values with median() in specific Series (column).
df = pd.read_csv("dirtydata.csv")
print(df.to_string())

print("After replacing...")
calories_median = df["Calories"].median()
df["Calories"]=df["Calories"].fillna(calories_median)

print(df.to_string())
```

```python
# Replace empty values with mode() in specific Series (column).
df = pd.read_csv("dirtydata.csv")
print(df.to_string())

print("After replacing...")
calories_mode = df["Calories"].mode()[0]
df["Calories"]=df["Calories"].fillna(calories_mode)

print(df.to_string())
```

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace = True)
print(df.to_string())
```

```python
# Histogram.
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data_pandas_plot.csv")
df["Duration"].plot(kind = "hist")

plt.show()
```

```python
#describe()  quick stats
print(df.describe())
```

```python
#apply()  custom column logic
df["Intensity"] = df["Pulse"].apply(
    lambda x: "High" if x > 120 else "Normal"
)
print(df[["Pulse", "Intensity"]].head())
```

```python
#groupby() average calories by duration
print(df.groupby("Duration")["Calories"].mean())
```

```python
#astype()  enforce numeric types
df["Calories"] = df["Calories"].astype(float)
```

```python
#sort_values()
df.sort_values(by="Calories", ascending=False, inplace=True)
print(df.head())
```
