```python
all_data = [['John','Amelie','Michael','Mary','Steven'],['Maria','Juan','Javier','Natalia','Pilar']]
#if we want a single list containeing all names with two or more e's in them use a for loop
names_of_interest=[]
for names in all_data:
    enough_es=[name for name in names if name.count('e')>=2]
    names_of_interest.extend(enough_es)

names_of_interest
```

```python
# we could use two list comprehensions instead of a for loop for the outer list above
flattened = [ x for List in all_data for x in List]
flattened
```

```python
# now follow same logic as above but add filter condition
result = [ x for List in all_data for x in List if x.count('e')>=2]
result
```
